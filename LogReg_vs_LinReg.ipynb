# BINARY CLASSIFIACTION : Logistic regression vs Linear regression

#### Why linear regression is not suitable for Binary Classification

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
lin_regression = LinearRegression()

from sklearn.linear_model import LogisticRegression
log_regression = LogisticRegression()

from sklearn.metrics import mean_squared_error, r2_score
```

```python
age = np.array([i for i in range(10,30)])
purchase = np.concatenate((np.zeros(10), np.ones(10)))
purchase
```

```python
purchase
```

```python
plt.plot(age, purchase, 'xb')  #'xb' stands for 'xbold' --> another options : '+'
plt.xlabel('AGE')
plt.ylabel('0 for NO Purchase | 1 for Purchase')
plt.grid(color='k', linestyle=':', linewidth=0.8)
plt.show()
```

```python
lin_regression.fit(pd.DataFrame(age), purchase)
lin_predict = lin_regression.predict(pd.DataFrame(age))

line_5 = age * 0 + .5

plt.plot(age, purchase, 'xb')
plt.plot(age, lin_predict, 'r')
plt.plot(age, line_5, 'g')
plt.xlabel('AGE')
plt.ylabel('0 for NO Purchase | 1 for Purchase')
plt.grid(color='k', linestyle=':', linewidth=0.8)
plt.show()
```

```python
test_age = np.array([i for i in range(10,35,5)])
test_purchase_lin = lin_regression.predict(pd.DataFrame(test_age))
test_purchase_lin

```

```python
log_regression.fit(pd.DataFrame(age), purchase)
purchase_predict = log_regression.predict_proba(pd.DataFrame(age))
log_purchase_predict = [item[1] for item in purchase_predict]

plt.plot(age, purchase, 'xb')
plt.plot(age, log_purchase_predict, 'r')
plt.plot(age, line_5, 'g')
plt.xlabel('AGE')
plt.ylabel('0 for NO Purchase | 1 for Purchase')
plt.grid(color='k', linestyle=':', linewidth=0.8)
plt.show()
```

#### Generally :   R --> The highest the better  |  RMSE --> The lowest the better

```python
print("Linear regression R :", r2_score(purchase, lin_predict))
print("Logistic regression R :", r2_score(purchase, log_purchase_predict))
print("Linear regression RMSE :", mean_squared_error(purchase, lin_predict))
print("Logistic regression RMSE :", mean_squared_error(purchase, log_purchase_predict))
```

```python
age = np.append(age, np.array([i for i in range(45,60)]))
purchase = np.append(purchase, np.ones(25))

```

```python
lin_regression.fit(pd.DataFrame(age), purchase)
lin_predict_2 = lin_regression.predict(pd.DataFrame(age))
line_5 = age * 0 + .5

plt.plot(age, purchase, 'xb')
plt.plot(age, line_5, 'g')
plt.plot(age, lin_predict_2, 'r')
plt.xlabel('AGE')
plt.ylabel('0 for NO Purchase | 1 for Purchase')
plt.grid(color='k', linestyle=':', linewidth=0.8)
plt.show
```

```python
test_age_2 = np.array([i for i in range(15,31)])
test_purchase_2 = lin_regression.predict(pd.DataFrame(test_age_2))
test_purchase_2
```

```python
log_regression.fit(pd.DataFrame(age), purchase)
purchase_predict_2 = log_regression.predict_proba(pd.DataFrame(age))
log_purchase_predict_2 = [item[1] for item in purchase_predict_2]

plt.plot(age, purchase, 'xb')
plt.plot(age, line_5, 'g')
plt.plot(age, log_purchase_predict_2,'r')
plt.xlabel('AGE')
plt.ylabel('0 for NO Purchase | 1 for Purchase')
plt.grid(color='k', linestyle=':', linewidth=0.8)
plt.show
```

```python
test_purchase_3 = log_regression.predict_proba(pd.DataFrame(test_age_2))
test_purchase_3
```

```python
print("Linear regression R :", r2_score(purchase, lin_predict_2))
print("Logistic regression R :", r2_score(purchase, log_purchase_predict_2))
print("Linear regression RMSE :", mean_squared_error(purchase, lin_predict_2))
print("Logistic regression RMSE :", mean_squared_error(purchase, log_purchase_predict_2))
```
